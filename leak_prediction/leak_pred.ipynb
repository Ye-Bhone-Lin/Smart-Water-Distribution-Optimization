```python
from scipy.stats import skew

num_cols = df.select_dtypes(include=['float64','int64']).columns

skewness = df[num_cols].apply(lambda x: skew(x.dropna()))
print(skewness)
```

```python
for col in num_cols:
    col_skew = skew(df[col].dropna())
    if abs(col_skew) < 0.5:  # normal-ish
        df[col].fillna(df[col].mean(), inplace=True)
    else:  # skewed
        df[col].fillna(df[col].median(), inplace=True)
```

##### How i solved the nan values. I visualized the Histogram and check the data behaviour whether normal distribution or left or right skewed. If the data behaviour has a right or left-skewed i chose medium instead of mean. If the data is normal distribution, i chose mean method. For this dataset, i didnt decide to drop the nan values because of the lack of data.

```python
import numpy as np

conditions = [
    df['class'] >= 95,
    df['class'] < 95
]

choices = ['Normal','Leak-Suspected']

df['class'] = np.select(conditions, choices, default='Unknown')
```

```python
import seaborn as sns 

sns.countplot(x='class', data=df)
```

```python
x = df.drop(['class','date'], axis=1)
y = df['class']
print(x.shape, y.shape)
```

##### we see that the target class is imblanced probably 98% and 2%. 

```python
from imblearn.over_sampling import SMOTE

smote=SMOTE(sampling_strategy='minority') 
x,y=smote.fit_resample(x,y)
y.value_counts()
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42, stratify=y)
print(x_train.shape, x_test.shape, y_train.shape, y_test.shape)
```

```python
import lightgbm as lgb
import shap

clf = lgb.LGBMClassifier(
    n_estimators=200,
    learning_rate=0.05,
    class_weight='balanced', # handle imbalance in target classes
    random_state=42
)
```

```python
clf.fit(x_train, y_train)
y_pred = clf.predict(x_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix

print("Classification Report:\n", classification_report(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

```python
explainer = shap.TreeExplainer(clf)
shap_values = explainer.shap_values(x_test)

shap.summary_plot(shap_values, x_test, plot_type="bar")
```

##### In that case we alredy see that PH-O is the most important feature in this dataset and secondly 

```python
y_train = y_train.map({'Normal':0, 'Leak-Suspected':1})
y_test = y_test.map({'Normal':0, 'Leak-Suspected':1})
```

```python
import xgboost as xgb

clf = xgb.XGBClassifier(
    n_estimators=200,
    learning_rate=0.05,
    random_state=42,
    use_label_encoder=False,
    eval_metric="logloss"
)

clf.fit(x_train, y_train)
```

```python
y_pred = clf.predict(x_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix

print("Classification Report:\n", classification_report(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
```
